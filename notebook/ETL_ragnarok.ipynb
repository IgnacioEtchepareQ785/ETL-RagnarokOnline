### Automatizar varios mobs


```python
monster_ids = [1004]
all_drops = []

for monster_id in monster_ids:
    url = f"https://ragnapi.com/api/v1/old-times/monsters/{monster_id}"
    response = requests.get(url)

    if response.status_code == 200:
        data = response.json()
        for drop in data["drops"]:
            drop["monster_id"] = data["monster_id"]
            drop["monster_name"] = data["monster_info"]
            all_drops.append(drop)


df_all = pd.DataFrame(all_drops)
df_all
```

```python
import requests
import pandas as pd
import time

monster_ids = [1001, 1002, 1003, 1004]
all_drops = []

headers = {
    "User-Agent": "Mozilla/5.0 (DataEngineerLearning)"
}

for monster_id in monster_ids:
    url = f"https://ragnapi.com/api/v1/old-times/monsters/{monster_id}"
    response = requests.get(url, headers=headers, timeout=10)

    print(f"Monster {monster_id} → Status {response.status_code}")

    if response.status_code != 200:
        print(f"⚠️ Monster {monster_id} no existe, se omite")
        continue

    if not response.text.strip():
        print(f"⚠️ Monster {monster_id} sin contenido")
        continue

    data = response.json()

    for drop in data.get("drops", []):
        all_drops.append({
            "monster_id": data["monster_id"],
            "monster_name": data["monster_info"],
            "item_name": drop.get("name"),
            "drop_rate": drop.get("rate")
        })

    time.sleep(0.5)  # evita bloqueos

df_all = pd.DataFrame(all_drops)
df_all
```
